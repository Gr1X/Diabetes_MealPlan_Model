```python
import matplotlib.pyplot as plt
import pandas as pd

# === Data hasil evaluasi dari masing-masing model ===
algorithms = [
    'Random Forest\n(Tuned)', 
    'XGBoost\n(Tuned)', 
    'CNN + LSTM\n(Tuned)', 
    'CNN + LSTM + SVM\n(Tuned)'
]

accuracy = [0.91, 0.9139, 0.8910, 0.89]
precision = [0.92, 0.92, 0.89, 0.89]
recall = [0.91, 0.91, 0.89, 0.89]
f1_score = [0.91, 0.91, 0.89, 0.89]

# === Membuat tabel perbandingan dalam DataFrame ===
comparison_df = pd.DataFrame({
    'Algoritma': algorithms,
    'Akurasi': accuracy,
    'Presisi (Weighted Avg)': precision,
    'Recall (Weighted Avg)': recall,
    'F1-Score (Weighted Avg)': f1_score
})

# === Menampilkan tabel ===
print("=== Tabel Perbandingan Kinerja Algoritma ===")
print(comparison_df.to_string(index=False))

# === Membuat Plot Perbandingan ===
plt.figure(figsize=(12, 8))

plt.scatter(algorithms, accuracy, label='Akurasi', color='skyblue', s=150, zorder=5)
plt.scatter(algorithms, precision, label='Presisi', color='salmon', s=150, zorder=5)
plt.scatter(algorithms, recall, label='Recall', color='lightgreen', s=150, zorder=5)
plt.scatter(algorithms, f1_score, label='F1-Score', color='plum', s=150, zorder=5)

# === Label dan Tampilan ===
plt.title('Perbandingan Kinerja 4 Algoritma', fontsize=16)
plt.ylabel('Nilai Metrik', fontsize=12)
plt.xlabel('Algoritma', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5)
plt.ylim(0.85, 0.95)

# === Menampilkan nilai pada setiap titik ===
for i, val in enumerate(accuracy):
    plt.annotate(round(val, 3), (i, val), textcoords="offset points", xytext=(0,10), ha='center', color='blue')
for i, val in enumerate(f1_score):
    plt.annotate(round(val, 3), (i, val), textcoords="offset points", xytext=(0,-20), ha='center', color='purple')

plt.xticks(rotation=10)
plt.tight_layout()
plt.show()
```
