```python
import matplotlib.pyplot as plt
import pandas as pd

# === Data hasil evaluasi dari masing-masing model ===
algorithms = [
    'Random Forest\n(Tuned)', 
    'XGBoost\n(Tuned)', 
    'CNN + LSTM\n(Tuned)', 
    'CNN + LSTM + SVM\n(Tuned)'
]

accuracy = [0.91, 0.9139, 0.8910, 0.89]
precision = [0.92, 0.92, 0.89, 0.89]
recall = [0.91, 0.91, 0.89, 0.89]
f1_score = [0.91, 0.91, 0.89, 0.89]

# === Membuat tabel perbandingan dalam DataFrame ===
comparison_df = pd.DataFrame({
    'Algoritma': algorithms,
    'Akurasi': accuracy,
    'Presisi (Weighted Avg)': precision,
    'Recall (Weighted Avg)': recall,
    'F1-Score (Weighted Avg)': f1_score
})

# === Menampilkan tabel ===
print("=== Tabel Perbandingan Kinerja Algoritma ===")
print(comparison_df.to_string(index=False))

# === Membuat Plot Perbandingan ===
plt.figure(figsize=(12, 8))

plt.scatter(algorithms, accuracy, label='Akurasi', color='skyblue', s=150, zorder=5)
plt.scatter(algorithms, precision, label='Presisi', color='salmon', s=150, zorder=5)
plt.scatter(algorithms, recall, label='Recall', color='lightgreen', s=150, zorder=5)
plt.scatter(algorithms, f1_score, label='F1-Score', color='plum', s=150, zorder=5)

# === Label dan Tampilan ===
plt.title('Perbandingan Kinerja 4 Algoritma', fontsize=16)
plt.ylabel('Nilai Metrik', fontsize=12)
plt.xlabel('Algoritma', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5)
plt.ylim(0.85, 0.95)

# === Menampilkan nilai pada setiap titik ===
for i, val in enumerate(accuracy):
    plt.annotate(round(val, 3), (i, val), textcoords="offset points", xytext=(0,10), ha='center', color='blue')
for i, val in enumerate(f1_score):
    plt.annotate(round(val, 3), (i, val), textcoords="offset points", xytext=(0,-20), ha='center', color='purple')

plt.xticks(rotation=10)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# === Data hasil evaluasi dari masing-masing model ===
algorithms = [
    'Random Forest\n(Tuned)', 
    'XGBoost\n(Tuned)', 
    'CNN + LSTM\n(Tuned)', 
    'CNN + LSTM + SVM\n(Tuned)'
]

accuracy = [0.91, 0.9139, 0.8910, 0.89]
precision = [0.92, 0.92, 0.89, 0.89]
recall = [0.91, 0.91, 0.89, 0.89]
f1_score = [0.91, 0.91, 0.89, 0.89]

# === Membuat DataFrame untuk kemudahan plotting ===
comparison_df = pd.DataFrame({
    'Algoritma': algorithms,
    'Akurasi': accuracy,
    'Presisi': precision,
    'Recall': recall,
    'F1-Score': f1_score
})

# === Plot dengan Bar Chart ===
plt.figure(figsize=(12, 7))

bar_width = 0.2
x = np.arange(len(algorithms))

plt.bar(x - 1.5*bar_width, accuracy, width=bar_width, label='Akurasi', color='#4F81BD')
plt.bar(x - 0.5*bar_width, precision, width=bar_width, label='Presisi', color='#C0504D')
plt.bar(x + 0.5*bar_width, recall, width=bar_width, label='Recall', color='#9BBB59')
plt.bar(x + 1.5*bar_width, f1_score, width=bar_width, label='F1-Score', color='#8064A2')

# === Label dan gaya tampilan ===
plt.xticks(x, algorithms, rotation=10, fontsize=11)
plt.ylabel('Nilai Metrik', fontsize=12)
plt.title('Perbandingan Kinerja Empat Algoritma (Tuned Models)', fontsize=16, pad=15)
plt.ylim(0.85, 0.95)
plt.legend(title='Metrik Evaluasi', fontsize=10)
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)

# === Menampilkan nilai di atas setiap batang ===
for i in range(len(algorithms)):
    plt.text(x[i] - 1.5*bar_width, accuracy[i] + 0.002, f"{accuracy[i]:.3f}", ha='center', fontsize=9, color='#2F5597')
    plt.text(x[i] - 0.5*bar_width, precision[i] + 0.002, f"{precision[i]:.3f}", ha='center', fontsize=9, color='#7F2418')
    plt.text(x[i] + 0.5*bar_width, recall[i] + 0.002, f"{recall[i]:.3f}", ha='center', fontsize=9, color='#385723')
    plt.text(x[i] + 1.5*bar_width, f1_score[i] + 0.002, f"{f1_score[i]:.3f}", ha='center', fontsize=9, color='#4B3863')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# === Data hasil evaluasi dari masing-masing model ===
algorithms = [
    'Random Forest (Tuned)', 
    'XGBoost (Tuned)', 
    'CNN + LSTM (Tuned)', 
    'CNN + LSTM + SVM (Tuned)'
]

accuracy = [0.91, 0.9139, 0.8910, 0.89]
precision = [0.92, 0.92, 0.89, 0.89]
recall = [0.91, 0.91, 0.89, 0.89]
f1_score = [0.91, 0.91, 0.89, 0.89]

# === Membuat DataFrame untuk kemudahan tampilan ===
comparison_df = pd.DataFrame({
    'Algoritma': algorithms,
    'Akurasi': accuracy,
    'Presisi': precision,
    'Recall': recall,
    'F1-Score': f1_score
})

# === Menampilkan tabel dengan Matplotlib ===
fig, ax = plt.subplots(figsize=(10, 2.5))
ax.axis('off')  # Hilangkan sumbu agar fokus ke tabel
ax.axis('tight')

# Membuat tabel dari DataFrame
table = ax.table(
    cellText=comparison_df.round(3).values,   # isi tabel
    colLabels=comparison_df.columns,          # header
    cellLoc='center',                         # rata tengah
    loc='center'
)

# === Styling tabel ===
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1.2, 1.5)  # memperbesar tabel agar proporsional

# Pewarnaan header agar lebih profesional
for (row, col), cell in table.get_celld().items():
    if row == 0:
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#4F81BD')  # warna biru header
    elif row % 2 == 0:
        cell.set_facecolor('#F2F2F2')  # warna abu muda untuk baris genap
    else:
        cell.set_facecolor('white')    # baris ganjil putih

plt.title("Tabel Perbandingan Kinerja Empat Algoritma (Tuned Models)", fontsize=13, pad=15)
plt.tight_layout()
plt.show()
```
